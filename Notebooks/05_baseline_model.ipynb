## 🎯 Objetivo

Implementar un modelo baseline que sirva como punto de referencia para evaluar el desempeño de los modelos de recomendación más avanzados desarrollados en las siguientes fases del proyecto.

El modelo baseline permitirá establecer una línea base de rendimiento utilizando una estrategia sencilla de recomendación, facilitando la comparación objetiva con los modelos Content-Based, Collaborative Filtering y el modelo híbrido.

---

## 📌 Objetivos específicos

- Construir un modelo de recomendación basado en popularidad.
- Generar recomendaciones iniciales para los usuarios.
- Evaluar el desempeño del modelo baseline.
- Establecer un punto de comparación para los modelos posteriores.

---

## 📂 Datos utilizados

- Dataset preparado en el Notebook 04.
- Variables generadas durante Feature Engineering.

---

## 📋 Actividades desarrolladas

- Carga del dataset preparado.
- Construcción del modelo baseline.
- Generación de recomendaciones.
- Evaluación inicial del modelo.
- Registro de resultados.

---

## 📈 Resultado esperado

Obtener un modelo baseline funcional capaz de recomendar productos utilizando una estrategia simple, cuyos resultados servirán como referencia para comparar el desempeño de modelos de recomendación más sofisticados.
